```python
# IMPORTANDO DAS BIBLIOTECAS NECESSÁRIAS
import pandas as pd # PARA MANIPULAÇÃO DE ARQUIVOS .CVS
from datetime import datetime # PARA MANIPULAÇÃO DE DATAS
```

```python
# CARRREGANDO DATA SET EM CSV PARA DATA FRAME
df = pd.read_csv('/content/public_servant_purchases_new.csv')
```

```python
# FUNÇÃO PARA DETECTAR COMPRAS DUPLICADAS
def detectar_compras_duplicadas(df):
  # USANDO A FUNÇÃO DUPLICATED() DO PANDAS PARA DETECTAR COMPRAS DUPLICADAS NAS COLUNAS
  duplicadas = df[df.duplicated(subset = ['nome_do_funcionario', 'item_comprado', 'valor_em_real'], keep=False)]
  return duplicadas
```

```python
detectar_compras_duplicadas(df) #TESTE
```

```python
# FUNÇÃO PARA DETECTAR COMPRAS COM VALORES SUSPEITOS (ACIMA DE R$1000.00)
def detectar_valores_suspeitos(df):
  # USANDO FILTROS EM NO DATA FRAME PARA RETORNAR VALORES MAIOR QUE R$100.00
  valores_suspeitos = df[df['valor_em_real'] > 1000.00]
  return valores_suspeitos
```

```python
detectar_valores_suspeitos(df) #TESTE
```

```python
# PREPARANDO DADOS: ADICIONA COLUNA DE DATA DE HORA SEPARADOS
# CONVERTENDO O VALOR DA COLUNA 'data_da_compra'PARA O FORMATO DATETIME
df['data_da_compra'] = pd.to_datetime(df['data_da_compra'])
# CRIANDO A COLUNA HORA NO FORMATO DESEJADO
df['hora'] = df["data_da_compra"].dt.strftime('%H:%M')
```

```python
df['hora'] #TESTE
```

```python
# FUNÇÃO PARA DETECTAR COMPRA FORA DO HORÁRIO COMERCIAL
def compras_fora_de_horario(df):
  # FILTRANDO A COLUNA HORA PARA RETORNAR OS DADOS DESEJADOS
  fora_de_horario = df[(df['hora'] < '08:00') | (df['hora'] > '18:00')]
  return fora_de_horario
```

```python
compras_fora_de_horario(df) #TESTE
```

```python
# FUNÇÃO PARA ORGANIZAR COMPRAS POR SERVIDOR EM DICIONÁRIO
def organizar_por_servidor(df):
  organizar_por_servidor = df.groupby('nome_do_funcionario')['item_comprado'].apply(list).to_dict()
  return organizar_por_servidor
```

```python
organizar_por_servidor(df) #TESTE
```

```python
# FUNÇÃO PARA GERAR RELATÓRIO
def gerar_relatorio(df):
    # RECEBENDO VALORES DAS FUNÇÕES
    duplicadas = detectar_compras_duplicadas(df)
    altos_valores = detectar_valores_suspeitos(df)
    fora_horario = compras_fora_de_horario(df)

    # JUNTADO FRAUDES ENCONTRADAS, EXCLUINDO DUPLICADOS
    total_fraudes = pd.concat([duplicadas, altos_valores, fora_horario]).drop_duplicates()

    # DEFINIDO SEVERIDADE
    severidade = 'Alta' if len(total_fraudes) > 10 else 'Média' if len(total_fraudes) > 5 else 'Baixa'

    # RELATÓRIO EM FORMATO DE DICIONARIO
    relatorio = {
        "total_fraudes": len(total_fraudes),
        "severidade": severidade,
        "duplicadas": len(duplicadas),
        "valores_suspeitos": len(altos_valores),
        "fora_do_horario": len(fora_horario),
        "compras_por_servidor": organizar_por_servidor(df)
    }

    return relatorio
```

```python
gerar_relatorio(df) #TESTE
```
